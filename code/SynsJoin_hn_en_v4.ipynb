```python
def getLines(file,splitby):
    with codecs.open(file, 'r', encoding="utf-8") as filer:
                    data = filer.read()
                    data = re.sub('\t+', '\t', data)
    lines = data.split(splitby)
    return lines
```

```python
def get_sentence(row):
    ret=re.findall('"(.+?)"', row['gloss'])
    return ret
def get_gloss(row):
    ret = []
    split = row['gloss'].split(';')
    for i in split:
        if '"' not in i:
            ret.append(i)
        else:
            break
    return ret
def getindexdefault(arr,elem, default):
        try:
            thing_index = arr.index(elem)
            return thing_index
        except ValueError:
            return default
```

###  Generating the final linked data set 
#### [hin_word, eng_conc]  from the concept-dict are used to filter the hin_eng linked dataset 

```python
con2 = WXC(order='utf2wx')
def get_fin_df(row,d,ign,c):
    conc_label_hn = row['conc_label_utf']
    conc_eng = row['conc_in_eng']
    mrs_conc = row['mrs_conc']
    hn_label =','+ conc_label_hn.split('_')[0]+','
    eng_label = conc_eng.split('_')[0]
    # checking if the hindi label is present in the hindi synset words list of linked_df
    #if the word is a verb
    if '_v_' in row['mrs_conc']:
        # adding 'ना' suffix before filtering
        hn_label_na = ','+conc_label_hn.split('_')[0] + 'ना' +','
        # adding 'करना' suffix before filtering
        hn_label_krna = ','+conc_label_hn.split('_')[0] + ' करना' + ','
        # adding 'होना' suffix before filtering
        hn_label_hona = ','+conc_label_hn.split('_')[0] + ' होना' + ','
        mask = linked_df.hindi_synset_words.swifter.progress_bar(False).apply(lambda x: (hn_label in x) or (hn_label_na in x) or (hn_label_krna in x) or (hn_label_hona in x))
    # if the word is not a verb
    else:
        mask = linked_df.hindi_synset_words.swifter.progress_bar(False).apply(lambda x: hn_label in x)

    
    #mask = linked_df.hindi_synset_words.apply(lambda x: (hn_label in x) or (hn_label_na in x) or (hn_label_krna in x))
    # checking if the eng label is present in the eng sysnet word list of linked_df
    mask1 = linked_df.synset_words.swifter.progress_bar(False).apply(lambda x: eng_label in x)
    # taking intersection of both the conditions
    cross = linked_df[mask & mask1]
    cross = cross.reset_index()  
    # if this pair of [hindi_label and conc_eng] are found in the same row of linked_df
    if cross.empty == False:
        
        for index, row in cross.iterrows():
            #if (getindexdefault(d['eng_id'],row['synset_id'],-1)!=getindexdefault(d['hin_id'],row['hindi_synset_id'],-1)):
            #if (row['synset_id'] not in d['eng_id']) and (row['hindi_synset_id'] not in d['hin_id']):
                # setting a flag signifying that the hindi word and conc eng pair were found in the same row
                #d['ign'].append(False)
                
                d['hin'].append(conc_label_hn)
                d['eng'].append(conc_eng)
                d['eng_id'].append(row['synset_id'])
                d['hin_id'].append(row['hindi_synset_id'])
                d['eng_gloss'].append(row['gloss'])
                d['eng_ex'].append(row['eng_ex'])
                if len(hn_syns_df[hn_syns_df['ID']==row['hindi_synset_id']]) != 0:
                    # getting hindi example and concept from the hindi synset dataframe
                    d['hin_ex'].append(hn_syns_df[hn_syns_df['ID']==row['hindi_synset_id']].EXAMPLE.item())
                    d['hin_gloss'].append(hn_syns_df[hn_syns_df['ID']==row['hindi_synset_id']].CONCEPT.item())
                else:
                    # if the hindi synset id is not found in hindi synset df. then adding empty strings
                    d['hin_ex'].append("")
                    d['hin_gloss'].append("")
                hn_syn_list = row['hindi_synset_words'].split(',')
                for word in hn_syn_list:
                    c['conc_label_utf'].append(word)
                    c['conc_label_wx'].append((con2.convert(word)))
                    c['conc_in_eng'].append(conc_eng)
                    c['mrs_conc'].append(mrs_conc)
    
    # else taking the corresponding hindi and english details(gloss,id, examples) separately in a many to many fashion
    else:
        
        # adding hindi-word and eng_concept pairs not found in the intersection to the ign dict
        ign['hindi_ig'].append(conc_label_hn)
        ign['eng_conc_ig'].append(conc_eng)
        ign['mrs_conc'].append(row['mrs_conc'])
        
                
```

```python
d = {'hin' : [],'eng' : [], 'hin_id':[],'eng_id':[], 'hin_gloss':[],'eng_gloss':[], 'hin_ex':[],'eng_ex':[]}
c = {'conc_label_wx':[],'conc_label_utf':[],'conc_in_eng':[],'mrs_conc':[]}
ign = {'hindi_ig' :[], 'eng_conc_ig':[],'mrs_conc':[]}
mrs_df.apply(lambda row: get_fin_df(row,d,ign,c), axis=1)
# final dataframe depending on the hindi label and eng conc from mrs
fin_df = pd.DataFrame.from_dict(d)
#ignored pairs which were not found in eng-hindi linked df
ign_df = pd.DataFrame.from_dict(ign)
```

```python
def count_mrs(row,h):
    if row['conc_label_wx'] in h.keys():
        h[row['conc_label_wx']] +=1
        row['conc_label_wx'] += '_' + str(h[row['conc_label_wx']])
        row['conc_label_wx'] = row['conc_label_wx'].replace(' ','+')
    else:
        h[row['conc_label_wx']] = 1
        row['conc_label_wx'] += '_' + str(h[row['conc_label_wx']])
        row['conc_label_wx'] = row['conc_label_wx'].replace(' ','+')

    return row['conc_label_wx']
```

### Checking if any yet unencountered eng synsets exist in the hin-eng linked dataset for ignored hindi words.

```python
def get_coarse_engsyn(row,d):
    conc_label_hn = row['hindi_ig']
    conc_eng = row['eng_conc_ig']
    hn_label =','+ conc_label_hn.split('_')[0]+','
    hn_label = hn_label.replace('+',' ')
    eng_label = conc_eng.split('_')[0]
    # checking ig the hindi label is present in the hindi synset words list of linked_df
    if '_v_' in row['mrs_conc']:
        # adding 'ना' suffix before filtering
        hn_label_na = ','+conc_label_hn.split('_')[0] + 'ना' +','
        # adding 'करना' suffix before filtering
        hn_label_krna = ','+conc_label_hn.split('_')[0] + ' करना' + ','
        # adding 'होना' suffix before filtering
        hn_label_hona = ','+conc_label_hn.split('_')[0] + ' होना' + ','
        mask = linked_df.hindi_synset_words.apply(lambda x: (hn_label in x) or (hn_label_na in x) or (hn_label_krna in x) or (hn_label_hona in x))
    # if the word is not a verb
    else:
        mask = linked_df.hindi_synset_words.apply(lambda x: hn_label in x)
    cross = linked_df[mask]
    if cross.empty == False:
        for index,row in cross.iterrows():
            d['hin'].append(conc_label_hn)
            d['eng_syns'].append(row['synset_words'])
            d['eng_id'].append(row['synset_id'])
            d['hin_id'].append(row['hindi_synset_id'])
            d['eng_gloss'].append(row['gloss'])
            d['eng_ex'].append(row['eng_ex'])
            if len(hn_syns_df[hn_syns_df['ID']==row['hindi_synset_id']]) != 0:
                # getting hindi example and concept from the hindi synset dataframe
                d['hin_ex'].append(hn_syns_df[hn_syns_df['ID']==row['hindi_synset_id']].EXAMPLE.item())
                d['hin_gloss'].append(hn_syns_df[hn_syns_df['ID']==row['hindi_synset_id']].CONCEPT.item())
            else:
                # if the hindi synset id is not found in hindi synset df. then adding empty strings
                d['hin_ex'].append("")
                d['hin_gloss'].append("")
                    
                
```
