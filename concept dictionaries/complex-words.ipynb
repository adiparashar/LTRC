```python
def getLines(file,splitby):
    with codecs.open(file, 'r', encoding="utf-8") as filer:
                    data = filer.read()
                    data = re.sub('\t+', '\t', data)
    lines = data.split(splitby)
    return lines
```

```python
def clean_wx(row):
    wx = row['conc_label_wx']
    wx = ''.join([i for i in wx if not i.isdigit()])
    wx = wx.replace('_','')
    wx = wx.replace('+',' ')
    return wx
def clean_en(row):
    en = row['conc_in_eng']
    en = ''.join([i for i in en if not i.isdigit()])
    en = en.replace('_','')
    en = en.replace('+',' ')
    return en
mrs_df['clean_wx_conc'] = mrs_df.apply(lambda x: clean_wx(x),axis=1)
mrs_df['clean_wx_en'] = mrs_df.apply(lambda x: clean_en(x),axis=1)
```

```python
new_nc = {}
NC_found = []
NC_new = {'hn_wx':[],'en':[]}
def check_nc(row):
    nc_wx = row['wx'][0].strip(' ')
    nc_en = row['en'][0]
    cross1 = mrs_df[mrs_df.clean_wx_conc.apply(lambda x : x == nc_wx)]
    if '[' in nc_wx:
        #print(row)
        nc_wx = nc_wx.replace('[','')
    if ']/CP' in nc_wx:
        nc_en = nc_en.replace(']/CP','')
        nc_en = nc_en.replace('.','')
    if cross1.empty == False:
        # adding found NCs in the dictionary to a list
        NC_found.append(nc_wx)
    else:
        # checking if the NC is multiword
        if ' ' in nc_wx:
            nc_l = nc_wx.split(' ')
            nc_n = ''
            # iterating over components of the NC for individual checks
            for i in nc_l:
                nc_n += i
                #print(i)
                cross1 = mrs_df[mrs_df.clean_wx_conc.apply(lambda x : x == i)]
                if cross1.empty == False:
                    # adding a suffix 'x' if the component's entry is there in the dictionary
                    nc_n += '_x+'
                    #print(nc_n,nc_wx)
                    #print(['ppp',cross1['conc_label_wx']])
                else:
                    # adding a suffix 'y' for unfound components
                    nc_n +=  '_y+'
                    NC_new['hn_wx'].append(i+'_y')
                    NC_new['en'].append(nc_en)
            NC_new['hn_wx'].append(nc_n[:-1])
            NC_new['en'].append(nc_en)
            
        else:
            NC_new['hn_wx'].append(nc_wx)
            NC_new['en'].append(nc_en)
        
NC.apply(lambda x: check_nc(x),axis =1)
```
