```python
df0.shape
df = df0[['linearization_wx','sentence_wx']]
df = df.rename(columns={"linearization_wx":"source_text", "sentence_wx":"target_text"})
df.head()
```

```python
df = df.applymap(str)
!pip install sklearn
```

```python
#!pip install sklearn
from sklearn.model_selection import train_test_split
train_df,test_df=train_test_split(df,test_size=0.2)
train_df.shape,test_df.shape
```

```python
!pip install simplet5
```

```python
from simplet5 import SimpleT5
```

```python
model = SimpleT5()
```

```python
import torch
torch.cuda.empty_cache()
```

```python
model.from_pretrained(model_type="mt5")
```

```python
#for finetuning.
#also set the path for output directory
model.train(train_df=train_df, eval_df=test_df, source_max_token_len=256, target_max_token_len=64, batch_size=1, max_epochs=12, use_gpu=True, outputdir = "/content/drive/MyDrive/LTRC" )
```

```python
model.load_model("mt5","/content/drive/MyDrive/LTRC/simplet5-epoch-10-train-loss-0.2209-val-loss-0.7112",use_gpu=True)
```

```python
test_df=pd.read_csv("/content/drive/MyDrive/LTRC/USR_lin_test_data.txt",sep="\t")
```

```python
test_df['predictions_mt5_wx'] = test_df.linearization_wx.apply(lambda x: model.predict(x))
```
